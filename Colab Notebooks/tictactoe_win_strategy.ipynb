```python
def mle(M,S):
    #where M is the number of observations and S is the number of distinct observations.
    lower, upper = S, -((M*S-2*S+1)//(S-M)) # Upper bound is an exercise for the code reviewer
    while upper != lower + 1:
        next = (lower + upper)//2
        if (next-1)**M > (next-S)*next**(M-1): # No floating point errors here
            lower = next
        else:
            upper = next
    return lower

mle(10**7, 70000)
```
