```python
def beats(a,b):
    a_n = 0
    b_n = 0
    for i in range(len(a)):
        for j in range(len(b)):
            if a[i]>b[j]:
                a_n+=1
            if a[i]<b[j]:
                b_n+=1
    return a_n > b_n

from itertools import combinations, product, permutations




def is_nontransitive(s):
    for p in combinations(s, 3):
        for a,b,c in permutations(p):
            if beats(a,b) and beats(b,c) and (not beats(a,c)):
                return True
            #print(a,b,c)
    return False

def check_n_of_pairs_beats(n):
    k = n-1
    combs = [p for p in combinations(list(range(1, n+1)), k)]
    out = 0

    for i in combs:
        for j in combs:
            if i == j: continue
            if beats(i, j): 
                out +=1
                #print(i,j)

    return out

terms = []
for i in range(3, 10):
    terms.append(check_n_of_pairs_beats(i))
    print(terms)
```

```python
from random import randint
from tqdm import tqdm 

def sample_cons():
    up_to_start = 50
    up_to_l = 20
    s = randint(1, up_to_start)
    t = randint(8, up_to_l)
    return tuple(s + i for i in range(t))

for i in tqdm(range(10**7)):
    t_s = [sample_cons() for _ in range(3)]
    if is_nontransitive(t_s):
        print(t_s)
        break
```
