```python
def random_polyomino_walk(n):
    from random import choice
    def near(x,y):
        return set(((x,y+1),(x+1,y),(x-1,y),(x,y-1),))
    t = (0,0)
    s = set((t,))
    r = 0
    while r < (n-1):
        sn = near(*t) - s
        if len(sn) == 0:
            t = choice(tuple(s))
        else:
            t = choice(tuple(sn))
            r+=1
            s.add(t)
    return s 

for i in range(10):
    P = random_polyomino_walk(100)
    T = extract_m_from_set_directly(P)

    h = has_a_hole(T)
    if h:
        #print(len(T), len(T[0]))
        print('\n[', ',\n  '.join(map(str, T)), ']')
```
