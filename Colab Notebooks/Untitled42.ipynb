```python
from random import randint

Bound_b = 7
Dim = 7
N = 10
cells = [tuple([randint(0, Bound_b) for j in range(Dim)]) for i in range(N)]

print(cells)
```

```python
from math import sqrt
dist_arr = [sqrt(k) for k in range(1, Dim+1)]
print(dist_arr)
```

```python
from math import floor
n_cells = 3**Dim - 1
dt = 0.1
rule = [[floor(n_cells*(3 - dt)/8), floor(n_cells*(3 + dt)/8)],
        [floor(n_cells*(2 - dt)/8), floor(n_cells*(3 + dt)/8)]]
print(n_cells)
print(rule)
print(rule[0][1] - rule[0][0])
```
