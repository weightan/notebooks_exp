```python
#idea: "additive primes" (every natural number is 1, an additive prime or the sum of additive primes in exactly one way)
from itertools import chain, combinations

def powerset(s):
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    #s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(len(s)+1))

primes = {2}

for i in range(13):
    t = 2
    while t in set(map(sum, powerset(primes))):
        t+=1
    primes.add(t)

print(sorted(list(primes)))

```

```python
from gmpy2 import isqrt
from sympy import divisors, proper_divisors, prod, primefactors, factorint
from itertools import combinations

is_ok = lambda i:  int(isqrt(i-1))**2 == i-1

def sieve(n):
    s  = [True]*n

    for i in range(2, n):
        if i**2 + 1 > n:
            break

        if is_ok(i):
            for j in range(2, n):
                if j*i >= n:
                    break
                if is_ok(j):
                    s[j*i] = False
    out = []
    for i in range(2, n):
        if is_ok(i) and s[i]:
            out.append(i)
    return out


        
p = sieve(60000)
print(p)
print([factorint(i) for i in p])



# semip = sorted([prod(i) for i in combinations(p, 2)])

# def most_common(lst):
#     return max(set(lst), key=lst.count)
# m = most_common(semip)

# print(m, semip.count(m))





```
