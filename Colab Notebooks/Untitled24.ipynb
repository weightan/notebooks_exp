```python
from sympy import primefactors 
from tqdm import tqdm
terms= []
for i in tqdm(range(2, 200_000)):
    p = primefactors(i)
    if len(p) != 1:
        p = list(map(lambda x: format(x), p ))
        if all(j in format(i,) for j in p):
            terms.append(i)

print(terms)
print(len(terms))
```

```python
import matplotlib.pyplot as plt
plt.plot(terms, ',')
plt.ylabel('some numbers')
plt.show()      
```

```python
from sympy import primefactors 
from tqdm import tqdm
from itertools import permutations
terms= []

for i in tqdm(range(1, 10**12)):
    p = primefactors(i)
    if len(p) != 1:
        p = list(map(lambda x: format(x, 'b'), p )) #set(''.join(p)) == set(format(i,'b')) and
        if all(j in format(i,'b') for j in p):
            if any(format(i,'b') == ''.join(t) for t in permutations(p)):
                terms.append(i)
                print('\n', terms)

print('\n', terms)
print(len(terms))
```
