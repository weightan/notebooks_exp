```python
from sympy import nextprime, prod, isprime
from tqdm import tqdm
terms = []
for N in tqdm(range(1, 10)):
    qen = [2,]

    for i in range(N-1): qen.append(nextprime(qen[-1]))
    # print(qen)
    while 1:
        t = 4*prod(qen) + 1
        if isprime(t):
            terms.append(t)
            print(qen)
            break
        qen.append(nextprime(qen[-1]))
        qen = qen[1:]
print(terms)
```

```python
import matplotlib.pyplot as plt
plt.plot(terms, 'ro')
plt.ylabel('some numbers')
plt.yscale('log')
plt.show()
```

```python
from random import choice

s = [str(choice((0,1))) for i in range(100000)]
s = ''.join(s)
N = 5
st = '1'*N

def find_all(s, sub):
    return [i for i in range(len(s)) if s.startswith(sub, i)]

out = [0, 0]

occ = find_all(s, st)
for i in occ:
    if i + N + 1 < len(s):
        if s[i + N +1 ] == '0':
            out[0] +=1
        else:
            out[1] +=1
print(out)



```
