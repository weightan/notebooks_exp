```python
def recaman(n):

  seq = []
  appiers = set()

  for i in range(n):
    
    if i == 0:
         x = 0
    else:
         x = seq[i-1] - i 

    if x >= 0 and x not in appiers :
         seq.append(x)
         appiers.add(x)
    else:
         seq.append(seq[i-1] + i)
         appiers.add(seq[i-1] + i)
    #print(seq)

  return seq

print(recaman(30))
```
