```python
lines = ''
with open('data1.txt') as f:
    lines = list(map(lambda it:it[:-1], f.readlines()))
lines[-1] = '1'

data = []
for i in range(0, len(lines), 2):
    t = list(map(int, lines[i].split(' x ')))
    n = int(lines[i+1])

    print(t, n)
    data.append((t,n))
```

```python
from math import log

maxx = max(data, key = lambda it: it[0][0])[0][0]
maxy = max(data, key = lambda it: it[0][1])[0][1]
print(maxx, maxy)


mat = [[0]*17 for i in range(17)]

for t, n in data:
    mat[t[0]-1][t[1]-1] = log(n+2)
    mat[t[1]-1][t[0]-1] = log(n+2)

print(mat)

```
