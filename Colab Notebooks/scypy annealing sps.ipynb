```python
from itertools import combinations, permutations, chain
from random import shuffle, sample  

gen_n = 8
test_s = '1234512413524'

def search_for_permut(s, target_table, target_lenght):
    q = [0]*(gen_n+1)
    current_sum = 0
    for i in range(len(s)):
        ind = int(s[i])-1
        if target_table[ind] == 1 and q[ind] == 0:
            q[ind] = 1
            current_sum += 1
            if current_sum == target_lenght: return True
        elif target_table[ind] == 1:
            q = [0]*(gen_n+1)
            q[ind] = 1
            current_sum = 1
        else:
            q = [0]*(gen_n+1)
            current_sum = 0
    return False

def powerset(iterable):
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(1, len(s)+1))

start_bucket = [ ''.join(map(str, i)) for i in  powerset(range(1, gen_n+1)) if  len(i) > 2]
test_table = lambda arr_i: [1 if str(i) in arr_i else 0 for i in range(1,gen_n+1)]
test_dict = {i: tuple(test_table(i)) for i in start_bucket}

target = len(start_bucket)

def conc_or(a, b):
    for i in range(len(b),-1,-1):
        if a[len(a)-i:]==b[:i]:
            return a[:len(a)-i]+b
    return a + b

def eval_sps(sps, q):
    c = 0
    for i in q:
        if not search_for_permut(sps, test_dict[i], len(i)):
            if len(i) > 5:
                c += 2
            else:
                c += 1
    return c

def eval_sps_normal(sps, q):
    c = 0
    for i in q:
        if not search_for_permut(sps, test_dict[i], len(i)):
            c += 1
    return c

def eval_sps_2(sps, q):
    c = 0
    for i in q:
        if not search_for_permut(sps, test_dict[i], len(i)):
            c += len(i)
    return  c


from math import floor

def ground(arr):
    out = []
    for i in arr:
        if i< 1:
            out.append(str(1))
            continue
        if i > gen_n:
            out.append(str(gen_n))
            continue
        out.append(str(round(i)))
    return ''.join(out)




print(eval_sps_normal('1234512413524', start_bucket), target)
```

```python
import numpy as np
from scipy.optimize import dual_annealing

len_sps = 75

func = lambda x: eval_sps(ground(x), start_bucket)
lw = [0.9] * len_sps
up = [gen_n+0.1] * len_sps


def callb(x, f, context):
    if context == 0:
        if f == 0: print('horay!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!')
        print(ground(x), f)

for i in range(100):

    ret = dual_annealing(func, bounds=list(zip(lw, up)), maxiter = 1800, initial_temp = 5230, maxfun = 11_000, visit = 2.9, callback=callb)
    print('')
    
```
