```python
def print_m(M):
    print('[', ',\n  '.join(map(str, M)) , '],',)

def generate_random_table(n):
    from random import randint
    return [[randint(0, n-1) for j in range(n)] for i in range(n)]

def transpose(m):
    return [ [m[j][i] for j in range(len(m))] for i in range(len(m))]

print_m(generate_random_table(7))
```

```python

def check_neutral(g):
    h = len(g)
    for i in range(h):
        if all(g[i][j] == j for j in range(h)):
            return True
    return False

def check_assot(g):
    from itertools import combinations_with_replacement
    for a,b,c in combinations_with_replacement(range(len(g))):
        if g[g[a][b]][c] != g[a][[b][c]]:
            return False
    return True


```
