```python
from math import sqrt
dist = lambda a,b: sqrt(sum((a[i] - b[i])**2 for i in range(len(a))))

def split_arr(arr, func):
    out = []
    buffer_t = []
    for i in range(len(arr)):
        if func(arr[i]):
            out.append(buffer_t)
            buffer_t = []
        else:
            buffer_t.append(arr[i])
    out.append(buffer_t)
    return [i for i in out if len(i)>0]

def find_closest(a, s):
    min_d = 100000
    eps = 0.000001
    out = []

    for i in s:
        d = dist(i, a)
        if d<min_d and d>eps: min_d = d
    
    for i in range(len(s)):
        if min_d - eps < dist(s[i], a) < min_d + eps:
            out.append(i)
    return out

def parse_file(lines):
    lines = [i[:-1] for i in lines if len(i) > 2]
    chunks = split_arr(lines, lambda t: t[0] == '#')[1:3]
    import re
    numeric_const_pattern = '[-+]? (?: (?: \d* \. \d+ ) | (?: \d+ \.? ) )(?: [Ee] [+-]? \d+ ) ?'
    rx = re.compile(numeric_const_pattern, re.VERBOSE)
    chunks[0]  =[[float(t) for t in rx.findall(s)]  for s in chunks[0]]
    chunks[1]  =[[int(t) for t in rx.findall(s)][1:] for s in chunks[1]]

    return chunks[0], chunks[1] 


def get_shell(vert):
    edges = set()

    for i in range(len(vert)):
        t = find_closest(vert[i], vert)
        for v in t:
            edges.add(tuple(sorted([i,v])))
    
    return list(edges)


def faces_to_edg(s):
    out = set()

    for t in s:
        for j in range(1,len(t)):
            out.add(tuple(sorted([t[j-1], t[j]])))
        out.add(tuple(sorted([t[0], t[len(t)-1]])))
    return list(out)

def uplift_edg(s, t = 20):
    out = [(s[i][0]+t, s[i][1]+t) for i in range(len(s))]
    glue = [(i, i + 12) for i in range(t)]
    
    #print(glue)

    out.extend(glue)
    return out
```
