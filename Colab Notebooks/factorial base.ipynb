```python
from math import factorial
def to_factorial_s(n):
    out = []
    c = 1
    if n == 0:
        return [0]
    while n > 0:
        out.append(n%c)
        n = n//c 
        c +=1
    return out #[::-1]

def from_fact_s(s):
    #s = s[::-1]
    out = 0
    for i in range(len(s)):
        out  = out + s[i]*factorial(i)
    return out
```

```python
from sympy import prod
from functools import lru_cache
from tqdm import tqdm

# @lru_cache(maxsize=3200)
# def factb(n, p=2): return n if n<p else factb(n//p, p+1)*10 + n%p


def fact_mul_per(n, step = 0):
    if n == 1 or n == 0:
        return step
    m = prod(to_factorial_s(n)[1:])
    return fact_mul_per(m, step + 1)

#@lru_cache(maxsize=3200)
def fact_add_per(n, step = 0):
    if n == 1 or n == 0:
        return step
    m = sum(to_factorial_s(n)[1:])
    return fact_add_per(m, step + 1)


terms = [0]*40

for i in range(0 , 10**9):
    t = fact_add_per(i)
    if terms[t] == 0:
        terms[t] = i
        print(terms)
```
