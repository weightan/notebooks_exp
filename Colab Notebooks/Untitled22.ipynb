```python
from itertools import product
states = []
tr = [0,1,1,0,1,1,1,0][::-1]
for i in product((0,1), repeat = 3):
    states.append(list(i))

transitions = []
for i in states:
    st1 = [1,] + i[:2]
    st0 = [0,] + i[:2]

    transitions.append((states.index(i), states.index(st1), states.index(st0), ))

for i in zip(tr, transitions):
    print(i[1], i[0])
```
