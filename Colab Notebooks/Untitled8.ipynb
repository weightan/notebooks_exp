```python
from random import randint, sample



# b 3
# d 0 1 4 5 6 7 8

# s 2 3
# rd 0 1 2 4 5 6 7 8 

n_of_cells = 100


def get_connections(n):
    c = [i for i in range(n)]
    con = [[] for i in range(n)]

    for i in range(n):
        opport = [j for j in c if len(con[j]) < 8]
        n_con = randint(4, 8)
        if n_con>len(opport): return []
        adj = sample(opport, n_con)
        for j in adj:
            con[i].append(j)
    
    return con

def check_possible_cells(c, con, p):
    states = [c, ]
    current_c = c.copy()
    for i in range(p):
        new_c = []
        all_c = 0
        #print(current_c)
        for j in range(len(current_c)):
            s = 0
            
            #print(len(con))

            for t in con[j]: s += current_c[t]

            if current_c[j] == 1 and (s == 2 or s == 3):
                new_c.append(1)
                all_c +=1
                continue
            else:
                new_c.append(0)
                continue

            if current_c[j] == 0 and s == 3:
                new_c.append(1)
                all_c +=1
                continue
            else:
                new_c.append(0)
                continue

        #print(new_c)

        if all_c < 3:
            return 'nope'

        
        states.append(new_c)
        current_c = new_c

    return states

from tqdm import tqdm        

p = 3
for i in tqdm(range(1000000)):
    cells = [randint(0,1) for i in range(n_of_cells)]
    con = get_connections(n_of_cells)
    if con != []:
        out = check_possible_cells(cells, con, p)

        if out == 'nope':
            continue

        if out[0] == out[1]:
            continue

        if out[0] == out[-1]:
            print(out)
            print('good')
```
