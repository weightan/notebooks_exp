```python
from random import randint, random, choice


def baseline(my_history, my_resources):
    placed_s = [i[1] for i in my_history if i[0] == 'S']
    sus_s =  [i[1] for i in my_history if i[0] == 'H']
    targets =((0, 1, 2), (3, 4, 5), (6, 7, 8), (0, 3, 6), (1, 4, 7), (2, 5, 8), (0, 4, 8), (2, 4, 6))
    sum_in = lambda t: (t[0] in placed_s)+(t[1] in placed_s)+(t[2] in placed_s)
    is_possible_win_for_other = lambda t:  placed_s.count(t[0])< sus_s.count(t[0]) and placed_s.count(t[1]) < sus_s.count(t[1])  and placed_s.count(t[2]) < sus_s.count(t[2])

    if len(placed_s) == 0:
        return ('S', randint(0, 7))
    else:
        max_targets = [i for i in targets if sum_in(i) == 3]
        better_max_targets = [i for i in targets if placed_s.count(i[0]) > sus_s.count(i[0]) and placed_s.count(i[1]) > sus_s.count(i[1])  and placed_s.count(i[2]) > sus_s.count(i[2])]

        if len(better_max_targets) != 0:return ('C', targets.index(choice(better_max_targets)))  

        disrupt_these = [ i for i in targets if is_possible_win_for_other(i)]

        if len(disrupt_these) != 0: 
            if random()< 0.6 and my_resources[1] > 0:
                return ('S', choice(choice(disrupt_these)))
            if my_resources[0] > 0:
                return ('B', choice(choice(disrupt_these)))

        if random() < 0.9:
            if my_resources[1] != 0:return ('S', randint(0, 8))
            if my_resources[0] != 0:return ('B', randint(0, 8))
        else:
            if my_resources[0] != 0:return ('B', randint(0, 8))
            if my_resources[1] != 0:return ('S', randint(0, 8))

    return ('C', randint(0, 7))

def randombot(my_history, my_resources):
    placed_my = [i[1] for i in my_history if i[0] == 'S']
    placed_other = [i[1] for i in my_history if i[0] == 'H']
    targets =((0, 1, 2), (3, 4, 5), (6, 7, 8), (0, 3, 6), (1, 4, 7), (2, 5, 8), (0, 4, 8), (2, 4, 6))
    sum_in = lambda t: (t[0] in placed_my)+(t[1] in placed_my)+(t[2] in placed_my)
    guaranteed = lambda t:  placed_my.count(t[0])>placed_other.count(t[0]) and placed_my.count(t[1])>placed_other.count(t[1]) and placed_my.count(t[2])>placed_other.count(t[2])
    if len(placed_my) == 0:
        return ('S', randint(0, 8))
    else:
        max_targets = [i for i in targets if guaranteed(i)]
        if len(max_targets) != 0: return ('C', targets.index(max_targets[0])) 
        questionable_targets = [i for i in targets if sum_in(i)==3]
        if  len(questionable_targets) != 0 and random()<0.1: return ('C', targets.index(choice(questionable_targets))) 
        if random()<0.5:
            if my_resources[1] != 0:return ('S', randint(0, 8))
            if my_resources[0] != 0:return ('B', randint(0, 8))
        else:
            if my_resources[0] != 0:return ('B', randint(0, 8))
            if my_resources[1] != 0:return ('S', randint(0, 8))
    return ('C', randint(0, 7))

def randombot_suspicious(my_history, my_resources):
    placed_s = [i[1] for i in my_history if i[0] == 'S' ]
    sus_s =  [i[1] for i in my_history if i[0] == 'H']

    targets = ((0, 1, 2), (3, 4, 5), (6, 7, 8), (0, 3, 6), (1, 4, 7), (2, 5, 8), (0, 4, 8), (2, 4, 6))
    
    sum_in = lambda t: (t[0] in placed_s)+(t[1] in placed_s)+(t[2] in placed_s)
    sum_in_sus = lambda t: (t[0] in sus_s)+(t[1] in sus_s)+(t[2] in sus_s)

    if len(my_history) == 0:
        return ('S', randint(0, 8))
    else:
        max_target = max(targets, key = sum_in)
        better_max_targets = [i for i in targets if placed_s.count(i[0]) > sus_s.count(i[0]) and placed_s.count(i[1]) > sus_s.count(i[1])  and placed_s.count(i[2]) > sus_s.count(i[2])]
        if len(better_max_targets) !=0: return ('C', targets.index(better_max_targets[0])) 
        if sum_in(max_target) == 3 and sum_in_sus(max_target) == 0:return ('C', targets.index(max_target)) 

        if my_resources[1] != 0:
            return ('S', randint(0, 8))
        if my_resources[0] != 0:
            return ('B', randint(0, 8))

    max_target = max(targets, key = sum_in)
    if sum_in(max_target) == 3:return ('C', targets.index(max_target)) 

    return ('C', randint(0, 7))


def randombot_baiting(my_history, my_resources):
    placed_s = [i[1] for i in my_history if i[0] != 'S']
    targets = ((0, 1, 2), (3, 4, 5), (6, 7, 8), (0, 3, 6), (1, 4, 7), (2, 5, 8), (0, 4, 8), (2, 4, 6))
    sum_in = lambda t: (t[0] in placed_s)+(t[1] in placed_s)+(t[2] in placed_s)

    if len(my_history) == 0:
        return ('S', randint(0, 8))
    else:
        max_target = max(targets, key = sum_in)
        if random() < 0.1 and sum_in(max_target) == 3: return ('C', targets.index(max_target)) 

        if random()<0.5:
            if my_resources[1] != 0:return ('S', randint(0, 8))
            if my_resources[0] != 0:return ('B', randint(0, 8))
        else:
            if my_resources[0] != 0:return ('B', randint(0, 8))
            if my_resources[1] != 0:return ('S', randint(0, 8))

    return ('C', randint(0, 7))


def randombot_just_clone(my_history, my_resources):
    placed_s = [i[1] for i in my_history if i[0] == 'S']
    sus_s =  [i[1] for i in my_history if i[0] == 'H']

    if len(sus_s) == 0: sus_s = [randint(0, 8),]

    targets = ((0, 1, 2), (3, 4, 5), (6, 7, 8), (0, 3, 6), (1, 4, 7), (2, 5, 8), (0, 4, 8), (2, 4, 6))
    
    sum_in = lambda t: (t[0] in placed_s)+(t[1] in placed_s)+(t[2] in placed_s)

    if len(my_history) == 0:
        return ('S', randint(0, 8))
    else:
        if my_resources[1] != 0:
            return  ('S', choice(sus_s))
        if my_resources[0] != 0:
            return ('B', choice(sus_s))

    max_targets = [i for i in targets if sum_in(i) == 3]
    if len(max_targets) != 0:
        return ('C', targets.index(choice(max_targets))) 

    return ('C', randint(0, 7))

def play_one_target(my_history, my_resources, my_state = 0):
    placed_s = sorted([i[1] for i in my_history if i[0] != 'H'])
    if len(placed_s) == 3:
        return ('C', 0)
    else:
        return ('S', len(placed_s))
```

```python
def print_table(t, h):
    from tabulate import tabulate
    print(tabulate(t, headers=h, tablefmt='orgtbl'))

bot_map = {
    0:baseline,
    1:randombot,
    2:randombot_suspicious
}

# bot_map = {
#     0:randombot_suspicious,
#     1:randombot,
#     2:play_one_target,
#     3:randombot_just_clone,
#     4:baseline
# }

# S_ymbol B_lank H_idden C_laim W_in L_ose

player_num=len(bot_map)
targets = ((0, 1, 2), (3, 4, 5), (6, 7, 8), (0, 3, 6), (1, 4, 7), (2, 5, 8), (0, 4, 8), (2, 4, 6))

def play_match_tic_tac(num1,num2):
    actual_board = [0]*9
    resources = [[5,5],[5,5]]
    history = [[], []]
    nums = [num1,num2]
    score = [0,0]
    for x in range(40):
        player = x%2
        other = (1+player)%2
        r = bot_map[nums[player]](history[player], resources[player])

        #logic here 
        if r[0] == 'S':
            assert resources[player][1] > 0
            actual_board[r[1]] += 1 if player == 0 else -1
            resources[player][1] -= 1
        elif r[0] == 'B':
            assert resources[player][0] > 0
            resources[player][0] -= 1
        elif r[0] == 'C':
            if player == 0:
                if all(actual_board[j] > 0 for j in targets[r[1]]):
                    score = [1,0] 
                else:
                    score = [0,1]
            if player == 1:
                if all(actual_board[j] < 0 for j in targets[r[1]]):
                    score = [0,1]
                else:
                    score = [1,0] 
        else:
            assert False 
        #logic here 

        history[player].append(r)

        if score[0] != 0 or score[1] != 0:
            return score

    return score

def tournament(bot_map, iters = 200, p = 100):
    from tqdm import tqdm 
    player_num=len(bot_map)
    scores = [0]*player_num
    for _ in tqdm(range(iters)):
        for x in range(player_num):
            for y in range(player_num):
                for t in range(p):
                    scorex, scorey = play_match_tic_tac(x,y)
                    scores[x] += scorex
                    scores[y] += scorey
    out = []
    for k, score in enumerate(scores):
        out.append((bot_map[k].__name__, score))
    out = sorted(out, key = lambda t: t[1], reverse = True)
    print('')
    print_table(out, ['bot name', 'score'])

tournament(bot_map, 200)
```
