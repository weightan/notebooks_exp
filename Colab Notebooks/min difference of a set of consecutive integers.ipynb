```python
import math
import numpy as np

def step(g):
    a = g>>2
    g &= 3
    g[1:] += a[:-1]
    g[:-1] += a[1:]
    g[:, 1:] += a[:, :-1]
    g[:, :-1] += a[:, 1:]
    g[0] += a[1]
    g[:, 0] += a[:, 1]

def nd(g, n):
    for y in range(0, n):
        for x in range(0, n):
            if g[y, x] > 3:
                return True
    return False

def main(i):
    size = math.ceil(math.sqrt(i)/2.7)+3
    grid = np.zeros((size, size), dtype=np.uint32)
    grid[0, 0] = i
    while nd(grid, size):
        for _ in range(100):
            step(grid)
    return grid




G = main(600_000)
#print(G)

import matplotlib.pyplot as plt

cmap = 'hot'
size = 4
plt.figure(num = None, figsize=(size, size), dpi=300)
plt.axis('off')
plot = plt.imshow(G, cmap = cmap)#, interpolation='bicubic' )
plt.show()
plt.close()
```
