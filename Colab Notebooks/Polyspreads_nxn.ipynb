```python
import numpy as np
from itertools import permutations, combinations
from random import choice, sample
from tqdm import tqdm
N = 4
s = range(N**2)

def trim_m(M):

    tempM = M.copy()
    for i in range(len(M)): 
        if all(j == 0 for j in M[i]):
            tempM.remove(M[i])
    M = np.rot90(np.array(tempM)).tolist()
    tempM = M.copy()
    for i in range(len(M)): 
        if all(j == 0 for j in M[i]):
            tempM.remove(M[i])

    return np.rot90(np.array(tempM), 3)

def tile_to_str(A):
    return ''.join(''.join(map(str, i)) for i in A)

def has_a_symmetry(M):
    Mc = M.copy()
    for i in range(4):
        M = np.rot90(M)
        if np.array_equal(np.fliplr(M), M) or np.array_equal(np.transpose(M) , M) or np.array_equal(np.rot90(M, 2), M): return True
    return False

tiles = []
d = dict()

t = 0

for i in tqdm(combinations(s, 2*N-1)):
    t+=1

    # if t> 70000:
    #     break

    tile = [1 if j in i else 0 for j in s]
    tile = np.array(tile)
    tile = trim_m(np.resize(tile, (N,N)).tolist())

    for i in range(5):
        tile = np.rot90(tile)
        if tile_to_str(np.fliplr(tile).tolist()) in d or tile_to_str(np.transpose(tile).tolist()) in d:
            break
    else:
        tiles.append(tile.tolist())
        d[tile_to_str(tile)] = True
    
# for i in tiles:
#     print(i, ',')
print(len(tiles))

for i in sample(tiles, 40):
    print(i, ',')

```

```python

from tqdm import tqdm
import numpy as np
from random import randint, choice, shuffle, sample


def generate_rotrefl(tile):
    import numpy as np
    out = []
    tile = np.array(tile)
    for j in range(6):
        tile = np.rot90(tile)
        if tile.tolist() not in out:out.append(tile.tolist())
        if np.fliplr(tile).tolist() not in out:out.append(np.fliplr(tile).tolist())
        if np.transpose(tile).tolist() not in out:out.append(np.transpose(tile).tolist())
    return out

def draw_from_tset(p):
    s = []
    #print(p)
    for i in range(len(p)):
        t = tiles[p[i][0]][0]
        for a,b in t:
            s.append((a + p[i][1],b + p[i][2], i+1))
    minx, maxx = min(s, key = lambda t:t[0])[0], max(s, key = lambda t:t[0])[0]
    miny, maxy = min(s, key = lambda t:t[1])[1], max(s, key = lambda t:t[1])[1]
    t = np.zeros((maxx-minx + 1 , maxy-miny + 1))
    for i in s: t[i[0]-minx][i[1]-miny] = i[2]

    return t

def has_a_symmetry(M):
    #Mc = M.copy()
    for i in range(4):
        M = np.rot90(M)
        if np.array_equal(np.fliplr(M), M) or np.array_equal(np.transpose(M) , M) or np.array_equal(np.rot90(M, 2), M): return True
    return False
    
def extract_m_from_tset(p):
    s = []
    for i in range(len(p)):
        t = tiles[p[i][0]][0]
        for a,b in t:
            s.append((a + p[i][1], b + p[i][2]))

    minx, maxx = min(s, key = lambda t:t[0])[0], max(s, key = lambda t:t[0])[0]
    miny, maxy = min(s, key = lambda t:t[1])[1], max(s, key = lambda t:t[1])[1]
    t = np.zeros((maxx-minx + 1 , maxy-miny + 1))

    for i in s:t[i[0]-minx][i[1]-miny] = 1
    return t

def get_coord_from_m(m):
    out = set()
    for i in range(len(m)):
        for j in range(len(m[i])):
            if  m[i][j] != 0:
                out.add((i,j))
    return out

def all_displacements(s):
    out = []
    for x,y in s:
        out.append(set(((a-x, b-y) for a,b in s)) )
    return out

def near_coord(s):
    out = set()
    for x, y in s:
        out.update(((x+1,y),(x,y+1),(x-1,y),(x,y-1),(x+1,y+1),(x-1,y-1),(x+1,y-1),(x-1,y+1)))
    return out - s

def generate_batch_sets(s):
    out = []
    for i in s:
        out.extend(all_displacements(get_coord_from_m(i)))
    out = [(i, near_coord(i)) for i in out]
    return out

def place_random_tile(s, nearS, p):
    for x,y in nearS:

        #for i in sample(range(len(tiles)), len(tiles)):
        for i in range(len(tiles)):
            t = tiles[i]
            temptile = set(map(lambda h: (h[0]+x, h[1]+y), t[0]))
            per = temptile & s
            if len(per) == 0:
                s = s | temptile
                p.append([i, x, y])
                return s, (t[1]|nearS)-(s), p
    return 0, 0, 0

def place_random_tile_moreconnection(s, nearS, p):
    maxconnect = 0 
    maxi = -1
    maxtile = set()
    for x,y in nearS:
        #for i in sample(range(len(tiles)), len(tiles)):
        for i in range(len(tiles)):
            t = tiles[i]
            temptile = set(map(lambda h: (h[0]+x, h[1]+y), t[0]))
            per = temptile & s
            connect = temptile & nearS

            if len(per) == 0:

                if len(connect) > maxconnect:
                    maxi = [i, x, y]
                    maxconnect = len(connect)
                    maxtile = temptile
    if maxi == -1:
        return 0, 0, 0
    else:
        s = s | maxtile
        p.append(maxi)
        return s, (tiles[maxi[0]][1] | nearS)-(s), p


################ parameters

tile = [[1,0,0],
        [0,1,1],
        [1,0,1],]

up_to_n_tiles = 20
more_than_n_tiles = 1
append_func = place_random_tile
#append_func = place_random_tile_moreconnection

################


tiles = generate_rotrefl(tile)
tiles = generate_batch_sets(tiles)

for i in tqdm(range(10**7)):
    shuffle(tiles)
    p = [[randint(0, len(tiles)-1), 0,0]]
    s = tiles[p[0][0]]
    nearS = s[1]
    s = s[0]
    cc = 0
    for i in range(up_to_n_tiles-1):
        s, nearS, p = append_func(s, nearS, p)
        #s, nearS, p = place_random_tile_moreconnection(s, nearS, p)

        if p == 0:
            break
        if i+1 > more_than_n_tiles and i%2 == 1 and has_a_symmetry(extract_m_from_tset(p)):
            print('\n', draw_from_tset(p), ',')
            print(draw_from_tset(p).tolist(), ',')
            cc = 1
            break

    if cc == 1:
        break

    


```
