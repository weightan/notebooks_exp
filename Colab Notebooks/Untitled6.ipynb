```python
from itertools import combinations_with_replacement
from functools import reduce

d = lambda a, b: (a * b  + (a+b))

def dec(n):
    out = []
    for a, b in combinations_with_replacement(range(0, n+1), 2):
        if d(a,b) == n:
            out.append((a, b))
    return out

def rec_dec(n):

    for a, b in combinations_with_replacement(range(1, n+1), 2):

        if d(a, b) == n:
            return rec_dec(a) + rec_dec(b)

    return [n]

            
def mul_d(s):
    return reduce(d, s)

        
p = [ i  for i in range(1, 200) if len(rec_dec(i)) == 1]   
print(p)
print([mul_d(p[:i]) for i in range(1, len(p))])            

#print([ (i,dec(i)) for i in range(1, 100) if len(dec(i)) == 1])

#print([ (i ) for i in range(1, 200) if all(h == 4 for h in rec_dec(i))])

```
