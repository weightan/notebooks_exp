```python

import numpy
from PIL import Image
#from numba import jit

width, height, frames = 48,48,1024

#jit
def check(slice):
	keepalive = [2,3] #CGOL rules
	revive = [3] # 23/3
	n = numpy.count_nonzero(slice[numpy.array([[True,True,True],[True,False,True],[True,True,True]])])
	return (n in keepalive) if slice[1,1] else (n in revive)

def iterate(src):
	src_padded = numpy.pad(src,((1,1),(1,1)))
	dst = numpy.zeros(src.shape,dtype=bool)
	for y in range(src.shape[0]):
		for x in range(src.shape[1]):
			dst[y,x] = check(src_padded[y:y+3,x:x+3])
	return dst

board = numpy.random.binomial(1,0.5,(height,width)).astype(bool)

old_boards = []
imgs = []
itr = frames
i = 0
for i in range(frames):
	if itr == i: break
	imgs.append(Image.fromarray(board.astype(numpy.uint8)*255,mode='L'))
	old_boards.insert(0,board)
	old_boards = old_boards[:3]
	board = iterate(board)
	if itr == frames:
		for j, old_board in enumerate(old_boards):
			if numpy.array_equal(old_board,board):
				print('Stopped prematurely')
				itr = i + (j*2) + 1
	print(f'{((i*100)/frames):.2f}%')

print('Saving...')
imgs[0].save(
	'conway.gif',
	save_all=True,
	append_images=imgs[1:],
	duration=20
)
```
