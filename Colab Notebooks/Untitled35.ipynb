```python
conclist = lambda s: int(''.join(map(str, s)))


d = ''
with open('oddity.txt', 'r') as g:
    d = g.read()

terms = []
for i in d:
    if i.isnumeric():
        terms.append(int(i))

sums = []

for i in range(6, len(terms)):
    # sums.append(sum(terms[i-5:i]))
    sums.append(conclist(terms[i-3:i]))

import matplotlib.pyplot as plt
plt.plot(sums, ',')
# plt.ylabel('some numbers')
plt.show()
```
