```python
from itertools import combinations
import numpy as np

# def near_point(x,y):
#     return set(((x+1,y),(x,y+1),(x-1,y),(x,y-1),(x+1,y+1),(x-1,y-1),(x+1,y-1),(x-1,y+1)))

def near_point(x,y):
    return set(((x+1,y),(x,y+1),(x-1,y),(x,y-1),))

# def near_point(x,y):
#     return set(((x+1,y+1),(x+1,y-1),))

# def near_point(x,y):
#      return set(((x+1,y+1),(x-1,y-1),(x+1,y-1),(x-1,y+1)))

def extract_m_from_set_directly(s):
    minx = min(s, key = lambda t:t[0])[0]
    maxx = max(s, key = lambda t:t[0])[0]
    miny = min(s, key = lambda t:t[1])[1]
    maxy = max(s, key = lambda t:t[1])[1]
    t = np.zeros((maxx-minx+1, maxy-miny+1), dtype = int)
    for i in s:t[i[0]-minx][i[1]-miny] = 1
    return t

def bounding_box_of_set(s):
    minx = min(s, key = lambda t:t[0])[0]
    maxx = max(s, key = lambda t:t[0])[0]
    miny = min(s, key = lambda t:t[1])[1]
    maxy = max(s, key = lambda t:t[1])[1]
    return (maxx-minx+1, maxy-miny+1)

def check_a_inside_b(a, b):
    a, b = min((a,b), key = lambda t:len(t)), max((a,b), key = lambda t:len(t))

def generate_rotrefl(tile):
    out = []
    tile = np.array(tile)
    for j in range(4):
        tile = np.rot90(tile)
        if tile.tolist() not in out:out.append(tile.tolist())
        if np.fliplr(tile).tolist() not in out:out.append(np.fliplr(tile).tolist())
        if np.transpose(tile).tolist() not in out:out.append(np.transpose(tile).tolist())
    return out

def get_coord_from_m(m):
    out = set()
    for i in range(len(m)):
        for j in range(len(m[i])):
            if  m[i][j] != 0:
                out.add((i,j))
    return out

N = 3
out = []
useless_sets  = []
maxN = N**2

# maxN = 5

def step(s, near_s, g = 0):
    b =  bounding_box_of_set(s)
    # print(b, len(s))
    if b[0] == N and b[1] == N:
        out.append(s.copy())
    if b[0] > N or b[1] > N:
        return
    if g >= maxN:
        # out.append(s.copy())
        return

    for x,y in near_s:

        ds = s.copy()
        ds.add((x,y))
        if ds not in useless_sets:
            dnear = (near_s | near_point(x,y)) - ds
            dnear = set([k for k in set(dnear) if k[0] >= 0 ]) #and k[1] >= 0
            useless_sets.append(ds)
            step(ds, dnear, g + 1)
    return

    
ms = set([(0,0),])


nears = near_point(0,0)
step(ms, nears)

tiles = []

for i in out:
    t = extract_m_from_set_directly(i)
    rt = t.tolist()
    if all( j not in tiles for j in generate_rotrefl(t)):
        tiles.append(rt)

tiles = sorted(tiles, key = lambda t: sum(sum(i) for i in t))[::-1]

for i in tiles:
    print(i, ',')

```
