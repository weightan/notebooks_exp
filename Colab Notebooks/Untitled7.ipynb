```python
from sympy import isprime
#pal10 = lambda n: str(n) == str(n)[::-1]
pal10 = lambda n: n == n[::-1]

def numberToBase(n, b):
    if n == 0:
        return [0]
    digits = []
    while n:
        digits.append(int(n % b))
        n //= b
    return digits[::-1]

to_b = lambda n, k: ''.join(map(str, numberToBase(n, k)))

def is_ok(n):
    b = to_b(n, 4)
    a = to_b(n, 5)
    if not pal10(b + a):
        return False
    if not pal10(a + b):
        return False
    if not isprime(int(a + b)):
        return False
    if not isprime(int(b + a)):
        return False
    return True

from tqdm import tqdm
terms = []
for i in tqdm(range(10**10)):
    if is_ok(i):
        terms.append(i)
        print(i)
print(terms)
```

```python
from sympy import isprime

pal10 = lambda n: str(n) == str(n)[::-1]

def seq(n):
    k = 1
    while isprime(k*(2*n+k-1)//2):k+=1
    return k


from tqdm import tqdm
terms = []
max_t = 0
for i in tqdm(range(2, 10**10)):
    t = seq(i)
    if t > max_t:
        max_t = t
        terms.append(t)
        print(i, t)

print(terms)
```
