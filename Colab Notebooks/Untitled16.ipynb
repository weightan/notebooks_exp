```python
tr = lambda s: f'({s[0]},{s[1]}),{s[2]} += ({s[3]},{s[4]})\\n'

lines = []
with open('d1.txt') as f:
    lines = f.readlines()
s = '(0,12)'
for i in lines:
    t = ''.join([j for j in i if j.isnumeric() or j == ',' or j == '+'])
    t = list(map(int, [k for k in t.replace('+',',').split(',') if k != '']))
    
    if len(t) == 5: 
        print(tr(t))
        s += tr(t)
print(s)
```
