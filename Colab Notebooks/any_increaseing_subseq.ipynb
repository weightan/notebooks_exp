```python
from math import factorial
from itertools import product

n = 7
a = []
b = [[[x, y] for x, y in product(range(2, n), range(2, n)) if (x * y) % n == z and x <= y] for z in range(1, n)]
print(b)
c = [[[1]]]
x = 0
y = 0
t = []

while len(c[-1]) > 0:
    c.append([])
    for i in c[-2]:
        for j in i:
            for k in b[j-1]:
                d = i + k
                d.remove(j)
                d.sort()
                if True not in [True in [False not in [m.count(i) <= d.count(i) for i in m] for m in l] for l in c]:
                    c[-1].append(d)
for i in c:
    a += i

for i in range(len(a)):
    t.append(i * [0])
    for j in range(i, len(a)):
        b = a[i] + a[j]
        b.sort()
        x2 = 0
        y2 = 0
        for k in range(len(a)):
            for l in range(k, len(a)):
                c = a[k] + a[l]
                c.sort()
                if b == c:
                    x2 += 1
                    y2 += int(prod([factorial(c.count(m)) / (factorial(a[k].count(m)) * factorial(c.count(m) - a[k].count(m))) for m in range(min(c), max(c) + 1)]))
        x = max(x, x2)
        y = max(y, y2)
        t[-1].append(y2)

m = max([len(str(t[i][j])) for i, j in product(range(len(t)), repeat=2)])
print(*a, sep="\n")

for i in t:
    for j in i:
        print(j, (m-len(str(j))) * " ", end="")
    print("")

print(x, y)
```
