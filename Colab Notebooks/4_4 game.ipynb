```python
t = 3513 
from itertools import combinations, product

digits = ['1','2','3','4','5']
allowed = []

for i in range(1, 5 + 1):
    for j in product(digits, repeat = i):
        if len(set(j)) == len(j):
            allowed.append(''.join(j))

print(allowed)
out = []
def get_non_intersecting(s,):
    current_d = set(list(''.join(s)))
    if 1 <= len(current_d) < 5:
        if sorted(s) not in out:
            out.append(sorted(s))

    if len(current_d) == 5:
        if sorted(s) not in out:
            out.append(sorted(s))
        return

    for i in allowed:
        if current_d.isdisjoint(set(list(i))):
            new_s = s.copy()
            new_s.append(i)
            get_non_intersecting(new_s)

for i in allowed:
    get_non_intersecting([i,])
```
