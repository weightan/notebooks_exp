```python
from functools import lru_cache
import sys
sys.setrecursionlimit(25000)
from math import floor

@lru_cache(maxsize = 40000)
def a(n):
    if n in (0,):
        return 1
    return sum(a(floor(n/(2**i))) for i in range(1, n) )

terms = [ a(i) for i in range(1, 3000)]

print(terms)
```

```python
import matplotlib.pyplot as plt
plt.plot(terms, ',')
plt.ylabel('some numbers')
plt.show()
```

```python
from tqdm import tqdm
def find_cycle(s):
    for i in range(2, len(s)//2 + 1):
        if s[ len(s)-2*i : len(s)-i] == s[len(s)-i:]:
            return  s[-i:]
    return []

res = []

for N in tqdm(range(2, 550)):
    terms = [0, 8]
    c = 0
    for i in range(30_000):
        
        terms.append(int(str(terms[-1]+terms[-2])[::-1] ) % (terms[-1]))  
        if i % 3000 == 0:
            k = len(find_cycle(terms))
            if k != 0:
                c = 1
                res.append(k)
                break
    if c == 0:
        res.append(0)
        

print(res)

import matplotlib.pyplot as plt
plt.plot(res, ',')
plt.ylabel('some numbers')
plt.show()
```
