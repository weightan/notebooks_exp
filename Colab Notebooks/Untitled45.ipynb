```python
from itertools import chain, combinations, permutations
from random import randint
from tqdm import tqdm

gen_n = 8
test_table = lambda arr_i: [1 if i in arr_i else 0 for i in range(gen_n+1)]

def powerset(iterable):
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(3, len(s)+1))

def all_perm_powerset(iterable):
    out = []
    for i in powerset(iterable):
        temp = []
        for j in permutations(i):
            temp.append(''.join(map(str, j)))
        out.append(temp)
    return out #[::-1]

def all_perm_powerset_tables(iterable):
    out = []
    for i in powerset(iterable):
        temp = []
        for j in permutations(i):
            temp.append(''.join(map(str, j)))
        out.append(temp)
    
    tables = tuple(map(test_table, [list(map( int, list(out[k][0]))) for k in range(len(out))]))

    return out, tables

def conc_or(a, b):
    for i in range(len(b),-1,-1):
        if a[len(a)-i:]==b[:i]:
            return a[:len(a)-i]+b
    return a + b

print(all_perm_powerset_tables(range(1, gen_n+1)))
```

```python
def detect_lin_g(s):
    for i in range(2, (len(s)//3)+1):
        a = s[-i:]
        b = s[-i*2:-i]
        c = s[-i*3:-i*2]
        if (len(a) == len(b) == len(c)):
            if all(a[j] - b[j] == b[j] - c[j] for j in range(i)) :
                return i
    return -1

detect_lin_g([4,4,4,4,1,2,3,6,7,8,11,12,1])
```
