```python
from math import log2, floor
terms = []

for i in range(0, 19):
    for j in range(1, 2**i):
        terms.append(i+1)
print(terms[:100])

import matplotlib.pyplot as plt
plt.plot(terms)
plt.plot([floor(len(format(i, 'b'))) for i in range(len(terms))])
plt.ylabel('some numbers')
plt.show()
```
