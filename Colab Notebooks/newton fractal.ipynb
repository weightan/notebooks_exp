```python
import matplotlib.pyplot as plt
import numpy as np
import math as m
from tqdm import tqdm
minx=-10
maxx=10
miny=-10
maxy=10
width=1000
height=1000

def func():
    global width
    global height
    global maxx
    global maxy
    global miny
    global minx
    result=[]
    for X in tqdm(range(0,width)):
        this=[]
        for Y in range(0,height):
            x=minx+((X/width)*(maxx-minx))
            y=miny+((Y/height)*(maxy-miny))
            g=complex(x,y)
            iter_=0
            try:
                while m.sqrt((g**3-1).real**2+(g**3-1).imag**2)>.001 and iter_<50000:
                    g=g-((g**3-1)/(3*g**2))
                if m.isclose(complex(g).real,1,rel_tol=.001):
                    this.append(0)
                elif m.isclose(complex(g).real,-.5,rel_tol=.001) and complex(g).imag<0:
                    this.append(1)
                elif m.isclose(complex(g).real,-.5,rel_tol=.001) and complex(g).imag>=0:
                    this.append(2)
            except:
                this.append(0)
        result.append(this)
    return result
data=np.array(func())
plt.imshow(data,interpolation='nearest',cmap='Accent')
plt.show()
```
