```python
import numpy as np
from itertools import product
import matplotlib.pyplot as plt

def generate_metanimos(n = 3):
    out = []
    rotations = []
    for i in product((0,1), repeat = n**2):
        if sum(i) == 5:
            t = np.array(i).reshape((n,n))
            c = 1
            temparr = []
            for j in range(4):
                t = np.rot90(t)
                temparr.append(np.copy(t))

                #print( t )
                if t.tolist() in out:
                    c = 0
                    break


            if c == 1 :
                out.extend([g.tolist() for g in temparr])
                rotations.append([g.tolist() for g in temparr])

    return out, rotations




def vizualize(p):
    L = len(p)
    p = list(map(np.array, p))
    r = L//2 
    up = p[:r]

    up2 = []
    for t in range(17*2 -1):
        if t % 2 == 0:
            up2.append(up[t//2])
        else:
            up2.append(np.zeros((3,3)))


    down  = p[r:]

    down2 = []
    for t in range(17*2 -1):
        if t % 2 == 0:
            down2.append(down[t//2])
        else:
            down2.append(np.zeros((3,3)))
    

    out = np.block([*up2,np.zeros((3,3)),*down2])
    cmap = 'copper'
    cmap = 'gist_gray'
    print(cmap)
    fig = plt.figure()
    fig.patch.set_facecolor('000000')

    size = 8
    plt.figure(num = None, figsize=(size, size), dpi=300)
    plt.axis('off') 
    plot = plt.imshow(out, cmap = cmap)#, interpolation='bicubic' )
    plt.show()
    plt.close()



p, r = generate_metanimos(3)
#vizualize(p)

print(len(r))
print(len(p))

# for  k, v in enumerate(p):
#     print(np.array(v)*(k+1))
```
