```python
from itertools import permutations

def conc_or(a, b):
    for i in range(len(b),-1,-1):
        if a[len(a)-i:]==b[:i]:return a[:len(a)-i]+b
    return a+b

def assemples(s):
    start = s[0]
    for i in range(1, len(s)):
        if s[i] not in start:
            start = conc_or(start, s[i])
    return start

def get_superstrings(s):
    if len(s) > 8:
        return [0,]
    out = set()
    for i in permutations(s):
        out.add(assemples(i))
    ml = min([len(i) for i in out])
    return [i for i in out if len(i) == ml]

from random import sample, shuffle, choice, randint

def gen_str(len_s):
    s = sample(['1', '2','3','4',], len_s)
    #s = [choice(['1', '2','3','4','5']) for i in range(len_s)]
    return ''.join(s)


def find_superpower_nstring():
    seen_stuff = set()

    from tqdm import tqdm
    for i in range(10_000_000):
        n_of_strings = 3 #randint(4, 6)
        n_of_char = 4 #randint(2, 3)
        test_s = [gen_str(n_of_char) for j in range(n_of_strings)]
        test_s = list(set(test_s))
        #print(len(test_s), test_s)

        r = get_superstrings(test_s)
        c = 0
        k = 0
        k_2 = 0

        lengths = []

        for j in range(20):
            lengths.append(len(r))
            if len(r) == 1: break
            #print(len(r), r)

            pr_r = r.copy()
            r = get_superstrings(r)
            r = sorted(r)

            if pr_r == r:
                c = -1
                break

            if r[0] == 0:
                k = 1
                break

            if len(lengths) >= 5 and all(g == lengths[-1] for g in lengths[-5:]):
                k_2 = 1
                break
            c+=1
        
        # if k == 1:
        #     print('eploded ', test_s)

        if c not in (0,1,2) and k == 0 : #and k_2 == 0:
            if str(lengths) + str(c) + str(k_2) not in seen_stuff:
                seen_stuff.add(str(lengths) + str(c)  + str(k_2))
                print('terminated: ' + str(not k_2), 'len: ' + str(c), test_s, 'iteration lengths: ' + str(lengths))

find_superpower_nstring()


# f =   ['3142', '1423', '2134', '4231']
# r = get_superstrings(f)
# print(len(r), r)
# r = get_superstrings(r)
# print(len(r), r)
# r = get_superstrings(r)
# print(len(r), r)
# r = get_superstrings(r)
# print(len(r), r)
# r = get_superstrings(r)
# print(len(r), r)
# # print(len(r), list(map(int, r[0])))
```
