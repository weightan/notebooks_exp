```python
from tqdm import tqdm

terms = [1,]
s = set(terms)
for j in tqdm(range(100_000)):
    t = -1
    for i in range(0, len(terms)-1):
        d = terms[-1] - terms[i]
        if d<0:continue
        if (d < t or t == -1) and d not in s: t = d
    if t == -1:
        new_r = terms[-1]*3
        terms.append(new_r)
        s.add(new_r)
    else:
        terms.append(t)
        s.add(t)
print(terms)
```

```python
import matplotlib.pyplot as plt
nt = terms[:]
size = 3
lw = 0.4
#plt.figure(num = None, figsize=(size*2, size), dpi=300)

plt.plot(nt, '.', markersize  = 0.002)
plt.plot([i     for i in range(len(nt))], linewidth  = lw)
plt.plot([i*2   for i in range(len(nt))], linewidth  = lw)


# k = 620
# plt.plot([(i-k)*2 if i>k else 0 for i in range(len(nt))])

plt.plot([2*i/3 for i in range(len(nt))], linewidth  = lw)
plt.ylabel('a(n)')
plt.xlabel('n')
#plt.yscale('log')
plt.show()
```

```python
from tqdm import tqdm

terms = [1,3,]
s = set(terms)
for j in tqdm(range(10000)):
    t = -1
    for i in range(0, len(terms)-1):
        d = terms[-1] - terms[i]
        if d<=0:continue
        if (d < t or t == -1) and d not in s: t = d
    if t == -1:
        new_r = terms[-1]+terms[-2] 
        terms.append(new_r)
        s.add(new_r)
    else:
        terms.append(t)
        s.add(t)
print(terms)
```
