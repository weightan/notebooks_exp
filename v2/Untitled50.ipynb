```python
from random import sample, choice
N = 12

get_m = lambda k: [[0]*k for i in range(k)]

def print_m(m):
    print('\n['+',\n'.join(map(str, m))+']\n')

def check_min_d(candidates, points, bound):
    out_arr = []
    for x, y in candidates:
        if any(abs(a-x) + abs(b-y) < bound for a,b in points):
            out_arr.append((x,y))
    return out_arr


def connected_tile(k):
    s = k.copy()
    def near(x,y): return ((x,y+1),(x+1,y),(x-1,y),(x,y-1),)
    def start_from_seed(x,y):
        if (x,y) not in s:
            return
        s.remove((x,y))
        for dx, dy in near(x, y): start_from_seed(dx, dy)
    start_from_seed(*s[0])
    return len(s) == 0

for t_N in range(3, 14):
    s_arr = [ (a,b) for a in range(t_N) for b in range(t_N)]

    count = 0
    from tqdm import tqdm 
    for t in tqdm(range(10**4)):

        test_m = get_m(t_N)
        p = []
        good_ones = s_arr.copy()

        for i in range(t_N):

            cx, cy = choice(good_ones)
            p.append((cx, cy))
            good_ones.remove((cx, cy))
            test_m[cx][cy] = 1
            good_ones = check_min_d(good_ones, p, t_N-i)


        if connected_tile(p):
            count +=1
    print(f'n {t_N}, count {count}' )
# print_m(test_m)


```

```python
def get_coord_from_m(m):
    out = set()
    for i in range(len(m)):
        for j in range(len(m[i])):
            if  m[i][j] != 0:
                out.add((i,j))
    return out

def get_neighbourhood(s):
    out = s.copy()
    def near(x,y):
        return set(((x,y+1),(x+1,y),(x-1,y),(x,y-1), (x+1,y+1),(x+1,y-1),(x-1,y+1),(x-1,y-1),))
    for i in s: out = out | near(*i)
    return out-s 

def connected_tile_mod(k):
    s = k.copy()
    def near(x,y): return ((x,y+1),(x+1,y),(x-1,y),(x,y-1),)
    def start_from_seed(x,y):
        if (x,y) not in s:
            return
        s.discard((x,y))
        for dx, dy in near(x, y): start_from_seed(dx, dy)
    start_from_seed(*tuple(s)[0])
    return len(s) == 0

def has_a_hole_2(inp_s):
    s_1 = get_coord_from_m(inp_s)
    #print(s_1)
    return not connected_tile_mod(get_neighbourhood(s_1))
```
