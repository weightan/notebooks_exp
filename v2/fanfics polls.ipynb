```python
import json

users = dict()

def strip_d(s):
    t = s[::-1].index('-') + 2
    return s[:len(s)-t]

def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]

def parse_poll(s, text):
    lines = list(filter(lambda d: d != '', s.split('\n')))
    
    new_lines = [lines[0],]

    for i in range(1,len(lines)):
        nl = lines[i]

        if lines[i][-1] == '%' or lines[i] == '0 %' or lines[i-1] == '0 %' or 'Публичный опрос' in lines[i] or '- скрыть' in lines[i] or 'Вы не голосовали - хочу голосовать'in lines[i] or lines[i].startswith('Проголосовали '):
            continue

        if lines[i-1][-1] == '%':
            nl = strip_d(nl)
        
        new_lines.append(nl)

    name_poll = new_lines[0]
    if name_poll == '?' or  name_poll == '  '  or  name_poll == '.' or  name_poll == '???' or name_poll == ' ' or name_poll == '  ':
        name_poll = text
        #name_poll = name_poll.replace('#опрос', ' ')
        name_poll = name_poll.replace('\n', ' ')
        name_poll = name_poll.strip()
    
    if name_poll == 'Ваши действия?':
        name_poll = text + name_poll
        #name_poll = name_poll.replace('#опрос', ' ')
        name_poll = name_poll.replace('\n', ' ')
        name_poll = name_poll.strip() 
    
    if name_poll == 'Сколько?':
        name_poll = text + name_poll
        #name_poll = name_poll.replace('#опрос', ' ')
        name_poll = name_poll.replace('\n', ' ')
        name_poll = name_poll.strip() 

    new_lines = new_lines[1:]
    # if len(new_lines)%2 == 1:
    #     print(s)

    #     print(new_lines)


    for t in chunks(new_lines, 2):
        ans = t[0]
        new_users = t[1].split(', ')

        res = name_poll + ' => ' + ans
        res = res.replace('/n', '')

        for j in new_users:
            if users.get(j, []) == []:
                users[j] = [res,]
            else:
                users[j] = users[j] + [res,]
    # for i in new_lines:
    #     print(i)


f = open('t3.json')
  
data = json.load(f)
#print('len ', len( data['table']))

t = 0
for i in data['table']:

    # if 'помогите выбрать' in  i['poll'].lower():
    #     print(t)
    # parse_poll(i['poll'])

    if 'Еловая иголка\n9 июля 2021' != i['data']:
        parse_poll(i['poll'], i['text'])

    # if t == 2:
    #     parse_poll(i['poll'])
    #     print('\n')
    #     print(i['poll'])

    t+=1
    # if t > 1000:
    #     break
  
f.close()


```
