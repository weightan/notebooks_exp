```python
import pysat.solvers
from pysat.card import CardEnc as encode
from random import shuffle
import numpy as np

def draw(arr):
    import matplotlib.pyplot as plt
    # cmaps =  ['binary', 'gist_yarg', 'gist_gray', 'gray', 'bone',
    #                     'pink', 'spring', 'summer', 'autumn', 'winter', 'cool',
    #                     'Wistia', 'hot', 'afmhot', 'gist_heat', 'copper']
    cmap = 'copper'
    size = 3
    plt.figure(num = None, figsize=(size, size), dpi=300)
    plt.axis('off')
    plot = plt.imshow(arr, cmap = cmap)
    plt.show()
    plt.close()

for nn in range(2, 30):

    dim = (nn, nn)
    N = dim[0]*dim[1]

    v = []
    t = 0
    d = dict()

    for i in range(dim[0]):
        for j in range(dim[1]):
            t += 1
            d[(i,j)] = t
            v.append((i,j))

    clauses = []
    topid = N

    for a,b in v:
        g  = d[(a,b)]

        temp = [(a+1,b),(a,b+1),(a-1,b),(a,b-1),]

        if a == dim[0]-1:
            temp.remove((a+1,b))
        if b == dim[1]-1:
            temp.remove((a,b+1))
        if a == 0:
            temp.remove((a-1,b))
        if b == 0:
            temp.remove((a,b-1))


        if len(temp) == 4:
            c = encode.equals(lits = [d[k] for k in temp], bound = 2, top_id = topid).clauses
            clauses.extend( [[-g,] + i for  i in c])
            topid += 4
        else:
            clauses.extend( [[-g,]])

    for Bound in range(N, 1, -1):
        tempClauses = clauses + encode.equals(lits= [i for i in range(1,N +1)], top_id = topid, bound = Bound).clauses

        solver = pysat.solvers.Cadical()
        for clause in tempClauses:
            solver.add_clause(clause)
        sat = solver.solve()

        if sat:
            model = solver.get_model()
            temp = [1 if i > 0 else 0 for i in model[:N]]
            print(nn, sum(temp))

            temp = np.array(temp)
            temp.resize(dim)
            print(temp)
            # draw(temp)
            break
        else:
            pass
            #print('unsat')
```
