```python
n = 3
def near_bro(a,b, t):
    #1 2
    #3 4
    if t==0: return (a+1, b), (a, b+1)
    if t==1: return (a-1, b), (a, b+1)
    if t==2: return (a+1, b), (a, b-1)
    if t==3: return (a-1, b), (a, b-1)

memo = [0,]

end_p = set([(0,0),
            (n,0),
            (0,n),
            (n,n)])

is_ok = lambda t: 0<=t<n

def dfs(pos, n = 0):
    if n!= 0 and set(pos) == end_p:
        memo[0] += 1
        return 

    new_positions = [[] for i in range(4)]
    for i in range(len(pos)):
        for x,y in near_bro(*pos[i], i):
            if is_ok(x) and is_ok(y):
                new_positions[i].append((x,y))
    
    print(new_positions)
     
            
dfs(end_p)


```

```python
#https://oeis.org/A084834

from functools import lru_cache

def isqrt(n):
    if n > 0:
        x = 1 << (n.bit_length() + 1 >> 1)
        while True:
            y = (x + n // x) >> 1
            if y >= x:
                return x
            x = y
    elif n == 0:
        return 0
    else:
        raise ValueError("square root not defined for negative numbers")

from sympy import isprime

def is_pow(n, k):
    while n%k == 0: n = n//k
    return n == 1

@lru_cache(maxsize=10**5)
def any_power(n):
    return any((is_pow(n, k) for k in range(2,1+n//2)))

terms = [2,]

s= set(terms)
sums= set(terms)

for i in range(100):
    t = 2
    while t in s or any_power(t) or any(any_power(j + t) for j in sums):
        t+=1
    s.add(t);terms.append(t) 
    sums = set(map(lambda k:k+t, sums))
    sums.add(t)
    #print(i)

print(terms)
```

```python
from itertools import permutations
def rotate(s, n): return s[n:] + s[:n]
def hashhh(s):

    ns = [rotate(s, i) for i in range(1, n+1)]
    ns2 = [i[::-1] for i in ns]
    ns.extend(ns2)
    out = []
    for t in ns:
        nt = [tuple(map(lambda k: (k+i)%n, t)) for i in range(1, n)]
        out.extend(nt)

    ns = list(set(out))
    
    out = []
    for t in ns:
        nt = [list(map(lambda k: (k+i)%n, t)) for i in range(1, n)]
        out.extend(nt)
    return tuple(sorted(out)[0])

for n in range(2, 9):
    d = [i for i in range(n)]

    g = set()
    for i in permutations(d):
        g.add(hashhh(i))

    print(len(sorted(list(g))))
```

```python
def A357261_list(maxn):
    A = [1,]
    g = 3
    y = 3
    for n in range(2, maxn+1):
        if y != g:
            A.append(n%g)
            y += 1
        else:
            if (n-g+A[-1])%g == 0:
                A.append(g)
                g += 1
            else:
                 A.append((n-g+A[-1])%g)
    return A[:n]
terms = A357261_list(1000000)
print(terms)
```
