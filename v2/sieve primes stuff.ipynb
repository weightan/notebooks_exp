```python
c = [7, 11, 19, 53, 43, 173, 67, 2477, 8803, 9173, 32323, 37123, 163, 74093, 170957, 360293, 679733, 2404147, 2004917, 69009533, 51599563, 155757067, 96295483, 146161723, 1408126003, 3519879677, 2050312613, 3341091163, 78864114883, 65315700413, 1728061733, 9447241877]


df = lambda i: (i + 1)//2 - 1
c = list(map(df, c))
c = c[3:]
print(c)


from math import gcd
from functools import reduce
from sympy import sieve

sieve.extend_to_no(100)
pr = list(sieve._list)
for k, v in enumerate(c):
    
 
    g = reduce(gcd, [t**v + 1 for t in pr[:k+3]])

    print(v, k+3, g)
```

```python
from sympy import sieve
from tqdm import tqdm
from functools import reduce

sieve.extend_to_no(50)

pr = list(sieve._list)

terms = [1]*14 +  [0]*100

for i in tqdm(range(300000, 339866 + 10)):

    k, g, len_f = 1, 2, 0

    while g != 1:

        k += 1

        len_f += 1

        g = reduce(gcd, [t**i + 1 for t in pr[:k]])

    if len_f > 1 and terms[len_f] == 0:

        terms[len_f] = i

        print(terms)
```
