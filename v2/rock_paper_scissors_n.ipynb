```python

def print_m(M):
    print('[', ',\n  '.join(map(str, M)) , '],\n',)


class Pair:
    def __init__(self, first, second):
        self.first = first
        self.second = second
        
infi = 10**7;

class Node:
    def __init__(self, vertexNumber):    
        self.vertexNumber = vertexNumber
        self.children = []

    def Add_child(self, vNumber, length):
        p = Pair(vNumber, length);
        self.children.append(p);
    

def dijkstraDist(g, s):
    dist = [infi for i in range(len(g))]
    visited = [False for i in range(len(g))]

    dist[s] = 0;
    current = s;
    sett = set()   

    while (True):
        # Mark current as visited
        visited[current] = True;
        for i in range(len(g[current].children)):
            v = g[current].children[i].first;        
            if (visited[v]):
                continue;

            # Inserting into the
            # visited vertex
            sett.add(v);
            alt = dist[current] + g[current].children[i].second;

            if (alt < dist[v]):    
                dist[v] = alt;

        if current in sett:        
            sett.remove(current);    
        if (len(sett) == 0):
            break;

        # The new current
        minDist = infi;
        index = 0;
        for a in sett:    
            if (dist[a] < minDist):        
                minDist = dist[a];
                index = a;        
        current = index;

    return dist;
```

```python
def create_graph(matrix):
    v = []
    dr = len(matrix)
    for i in range(dr):
        a = Node(i)
        for j in range(dr):
            if i !=j and matrix[i][j] != 0: a.Add_child(j, 1)
        v.append(a)
    return v

def get_full_m(n):
    return [ [1 if i!=j else 0 for i in range(n) ]  for j in range(n)] 

def get_empty_m(n):
    return [ [0 for i in range(n) ]  for j in range(n)] 


from random import sample, randint
from tqdm import tqdm


def get_random_m(n, population):
    pos = [(i,j) for i in range(n) for j in range(n)]
    pos = [i for i in pos if i[0] != i[1]]
    s = set(sample(pos, population))
    return [ [1 if (i,j) in s else 0 for i in range(n)]  for j in range(n)] 

def get_random_balanced_m(n, r):
    m = get_empty_m(n)

    get_col =  lambda c, m: [i[c] for i in m]

    for x in range(n):
        colum = m[x]
        s = sum(colum)
        if s < r:
            diff = [j for j in colum if j == 0 and j!= x]
            for y in sample(diff, r - s):
                m[x][y] = 1
    
    for x in range(n):
        colum = get_col(x, m)
        s = sum(colum)
        if s < r:
            diff = [j for j in colum if j == 0 and j != x]
            for y in sample(diff, r - s):
                m[y][x] = 1
    
    return m
        




    
def all_d_less_than_k(k, m):
    n = len(m)
    tar_graph = create_graph(m)

    for s in range(n):
        dist = dijkstraDist(tar_graph, s)
        if not all(v <= k for v in dist):
            return False
    return True


def all_d_less_than_k_but_finite(k, m):
    n = len(m)
    tar_graph = create_graph(m)

    for s in range(n):
        dist = dijkstraDist(tar_graph, s)
        if not all(v <= k for v in dist) or any(v == infi for v in dist):
            return False
    return True

def all_d_equal_k(k, m):
    n = len(m)
    tar_graph = create_graph(m)

    for s in range(n):
        dist = dijkstraDist(tar_graph, s)
        #print(dist)
        if not all(v == k or v == 0 for v in dist):
            return False
    return True

# n = 5
# M = get_random_m(n, 20)
# print_m(M)
# all_d_equal_k(2, M)

# K = 2
# n = 10

# minall = 10**7

# for i in tqdm(range(100000)):
#     mpop = randint(2*n, 4*n)
#     if mpop >= minall:
#         continue

#     M = get_random_m(n, mpop)

#     if all_d_less_than_k_but_finite(K, M) and mpop < minall:
#         print('\n', mpop)
#         print_m(M)
#         minall = mpop
```

```python
def generate_rotrefl(tile):
    import numpy as np
    out = []
    tile = np.array(tile)
    for j in range(6):
        tile = np.rot90(tile)
        if tile.tolist() not in out:out.append(tile.tolist())
        if np.fliplr(tile).tolist() not in out:out.append(np.fliplr(tile).tolist())
        if np.transpose(tile).tolist() not in out:out.append(np.transpose(tile).tolist())
    return out


def m_to_str(m):
    return ''.join(''.join(map(str, i)) for i in m)

def get_all_rot_str_m(m):
    return map(m_to_str, generate_rotrefl(m))
```
