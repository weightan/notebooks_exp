```python
from random import randint
from tqdm import tqdm
out = []
N = 10

for i in tqdm(range(10000)):
    g = [randint(0, N-1) for i in range(1000) ]

    s = 0
    for i in range(1, len(g)):
        if g[i-1] == g[i]:
            s+=1
    out.append(s)

import numpy as np

m = np.mean(out)
st = np.std(out)
print(m, st, min(out))
```
