```python
from random import randint, random, choice, sample, shuffle
from itertools import chain, combinations, permutations
#from numpy import exp
from tqdm import tqdm

def powerset(iterable):
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(1, len(s)+1 ))


N = 8

ps = tuple(powerset(range(1, N+1)))


per = [[''.join(map(str, p)) for p in permutations(i)] for i in ps if  2 <= len(i) <= 4]
#per = [sample(i, len(i)) for i in per]
print(len(per), sum(len(i) for i in per))

# per = per[::-1]
per = sample(per, len(per))

perstand = [[''.join(map(str, p)) for p in permutations(i)] for i in ps if len(i) >= 3]

L = len(per)
#Starter = ''.join(map(str, range(1, N+1)))
Starter = '12345'
#Starter = ''

def conc_or(a, b):
    for i in range(len(b),-1,-1):
        if a[len(a)-i:]==b[:i]:
            return a[:len(a)-i]+b
    return a + b

def assemble_sps(p):
    s = Starter
    for k,v in enumerate(p):
        c = 0
        for i in per[k]:
            if i in s:
                break
        else:
            s = conc_or(s, per[k][v])
    return s

def check_sups(s):
    c = 0
    for i in perstand:
        if not any(j in s for j in i):
            c+=1
    return c


def assemble_sps_permutations(perm,  M):

    prevgoodones = list(range(1, len(perm)))
    shuffle(prevgoodones) # = sample(prevgoodones, len(prevgoodones))
    s = choice(perm[prevgoodones[0]])
    prevgoodones = prevgoodones[1:]

    while True:
        if len(s) >=  M:
            return '0'
        goodones = []

        for i in prevgoodones:
            if len(perm[i][0]) > len(s) or not set(perm[i][0]).issubset(set(s[-10:])):
                goodones.append(i)
            else:
                for j in range(len(perm[i])):
                    if perm[i][j] in s:
                        break
                else:
                    goodones.append(i)

        if len(goodones) == 0:
            break

        minlen = 100000
        mins = 0
        mini = 0
        ok = False

        for i in goodones:
            c = False

            for j in range(len(perm[i])): 
                temps = conc_or(s, perm[i][j])
                lenT = len(temps)

                if lenT == len(s)+1:
                    s = temps
                    goodones.remove(i)
                    ok = True
                    c = True
                    break

                if lenT < minlen:
                    minlen = lenT
                    mins = temps
                    mini = i

            if c :
                break

        if not ok:
            s = mins
            goodones.remove(mini)
            
        prevgoodones = goodones

    return s

mins = ''
minlen = 760

count = 0

for i in tqdm(range(100000)):
    count += 1
    if count % 10==0:
        per = [sample(i, len(i)) for i in per]

    sps = assemble_sps_permutations(per, minlen)
    if sps != '0':
        clen = len(sps)
        t = check_sups(sps)
        v =  clen

        if v < minlen:

            minlen = v
            mins = sps

            print('len: ', clen, 'not included: ', t, 'sps: ', sps)


# n		a(n)
# 1		1
# 2		2
# 3		4
# 4		8
# 5		13
# 6		24
# 7		40
# 8		73..76
# 9		130..149

#'361582763412865371486358475213876428361273456187532641538274165276134875613245187264582317546'
#'624187543618257163824765831427563217358267185642378641723541738452163478213562376154827346852468'
```
