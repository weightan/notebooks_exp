```python
from math import gcd, factorial

def eval_s(s, k):
    f = lambda p: p**k  

    t = gcd(f(s[0]), f(s[1]))
    for i in range(2, len(s)):
        if t == 1:
            return i-2
        else:
            t = gcd(t, f(s[i]))
    return len(s)

def eval_s_exactly(s, k):
    t = gcd(s[0], s[1])
    for i in range(2, len(s)):
        if t == 1:
            return False
        else:
            t = gcd(t, s[i])
    return True

test_s = [i for i in range(2, 200)]
hits = [0]*210

from tqdm import tqdm


for k in tqdm(range(1, 10**6)):
    t = eval_s(test_s, k)

    if t<len(hits) and hits[t]==0:
        hits[t]=k
        #print(k)
        print(t, k, hits[1:])
```
